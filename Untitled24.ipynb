```python
image_folder = "images"
ids = []
embeddings = []
metadatas = []

if not os.path.exists(image_folder):
    print(f"WARNING: Folder '{image_folder}' does not exist")
else:
    print(f"Indexing images from {image_folder}...")
    files = [f for f in os.listdir(image_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

    if not files:
        print("No images found")

    for filename in files:
        path = os.path.join(image_folder, filename)
        try:
            vector = get_image_embedding(path)
            ids.append(filename)
            embeddings.append(vector)
            metadatas.append({"source": filename})
            print(f"Indexed: {filename}")
        except Exception as e:
            print(f"Error indexing {filename}: {e}")

    if ids:
        collection.add(embeddings=embeddings, ids=ids, metadatas=metadatas)
        print(f"Uploaded")
```

```python
import os
import chromadb
from PIL import Image
from transformers import CLIPProcessor, CLIPModel
import torch

print("Loading CLIP Model...")
model_id = "openai/clip-vit-base-patch32"
model = CLIPModel.from_pretrained(model_id)
processor = CLIPProcessor.from_pretrained(model_id)

print("Initializing ChromaDB...")
client = chromadb.Client()
try:
    client.delete_collection("food_gallery")
except:
    pass
collection = client.get_or_create_collection(name="food_gallery")

def get_image_embedding(image_path):
    image = Image.open(image_path)
    inputs = processor(images=image, return_tensors="pt")
    image_features = model.get_image_features(**inputs)
    image_features = image_features / image_features.norm(p=2, dim=-1, keepdim=True)

    return image_features.detach().numpy().flatten().tolist()

def get_text_embedding(text):
    inputs = processor(text=[text], return_tensors="pt", padding=True)
    text_features = model.get_text_features(**inputs)
    text_features = text_features / text_features.norm(p=2, dim=-1, keepdim=True)

    return text_features.detach().numpy().flatten().tolist()
```

```python
# @title
from IPython.display import display

query = "meat"

print(f"Searching for: '{query}'")
query_vector = get_text_embedding(query)

# Get Top 3 results to see what else it considered
results = collection.query(
    query_embeddings=[query_vector],
    n_results=3,
    include=["metadatas", "distances"]
)

print("\n--- RESULTS ---")
for i in range(len(results['ids'][0])):
    filename = results['ids'][0][i]
    score = results['distances'][0][i] # Distance (0 is identical, 1 is opposite)

    print(f"#{i+1}: {filename} (Distance: {score:.4f})")

    # Only show image if it's a reasonable match (e.g. distance ~ 1.5)
    display(Image.open(os.path.join(image_folder, filename)).resize((200, 150)))
```
